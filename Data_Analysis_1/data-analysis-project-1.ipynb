```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
print("complite")
```

```python
# Set visualization style
sns.set(style="whitegrid")
```

```python
# File path (update with your actual dataset name)
dataset_path = "/kaggle/input/global-population/Global _Population.csv"

# Load the dataset
data = pd.read_csv("/kaggle/input/global-population/Global _Population.csv")
```

```python
# Display the first few rows of the dataset
print("First 5 rows of the dataset:")
display(data.head())
```

```python
# Summary statistics
print("\nDataset Summary Statistics:")
display(data.describe())
```

```python
# Check for missing values
print("\nMissing Values in the Dataset:")
display(data.isnull().sum())
```

```python
# Data type information
print("\nData Types of Columns:")
display(data.info())
```

```python
# Visualize missing values (if any)
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Assuming 'data' is your dataset
# Filtering numerical columns only
numerical_columns = data.select_dtypes(include=[np.number])

if len(numerical_columns.columns) > 1:
    # Compute the correlation matrix for numerical features
    correlation_matrix = numerical_columns.corr()

    # Plot the heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Correlation Heatmap")
    plt.show()
else:
    print("Not enough numerical features for a correlation heatmap.")
```

```python
# Distribution of each numerical column
numerical_columns = data.select_dtypes(include=[np.number]).columns
for col in numerical_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], kde=True, bins=30, color="blue")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# Categorical column value counts
categorical_columns = data.select_dtypes(include=["object"]).columns

# Loop through each categorical column and print its value counts
for col in categorical_columns:
    print(f"\nValue Counts for {col}:")
    print(data[col].value_counts(dropna=False))  # dropna=False to include NaN counts
    print("-" * 50)  # Divider for better readability
```

```python
# Bar plot for categorical columns
categorical_columns = data.select_dtypes(include=["object"]).columns

for col in categorical_columns:
    plt.figure(figsize=(8, 4))  # Create a new figure for each column
    sns.countplot(
        x=data[col],  # Specify the column for the x-axis
        palette="viridis",  # Use the "viridis" color palette
        order=data[col].value_counts().index  # Order bars by value counts
    )
    plt.title(f"Count Plot for {col}")  # Title of the plot
    plt.xlabel(col)  # Label for x-axis
    plt.ylabel("Count")  # Label for y-axis
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
    plt.tight_layout()  # Adjust layout for clarity
    plt.show()  # Display the plot
```

```python
# Pairplot for relationships (numerical data)
numerical_columns = data.select_dtypes(include=[np.number]).columns

if len(numerical_columns) > 1:
    # Plotting the pairplot
    sns.pairplot(data[numerical_columns])
    plt.suptitle("Pairplot of Numerical Features", y=1.02)  # Adjust title position
    plt.show()
else:
    print("Not enough numerical features for pairplot visualization.")
```
